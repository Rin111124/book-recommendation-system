# Collaborative Filtering
User-based and item-based CF experiments with hyperparameter tuning hooks.

```python
import pandas as pd
from src import collaborative_filtering, data_preprocessing

raw = data_preprocessing.load_datasets()
books, users, ratings = data_preprocessing.clean_data(**raw)  # type: ignore[arg-type]
ratings = data_preprocessing.filter_interactions(ratings)

user_cf = collaborative_filtering.UserBasedCF()
user_cf.fit(ratings)
user_cf.recommend(user_id=276729, books=books, top_n=5)
```

```python
item_cf = collaborative_filtering.ItemBasedCF()
item_cf.fit(ratings)
item_cf.recommend(user_id=276729, books=books, top_n=5)
```
